```python

import random
from tempfile import tempdir
from typing import Counter
import matplotlib.pyplot as plt


#count=0
count1=0
count2=0
largest=1000
s=0


def randno(size, largest):
    random_no = []
    for _ in range(size):
        num = random.randint(1, largest)
        random_no.append(num)
    return random_no

def printlist(array):
    size = len(array)
    for x in range(size):
        print(x, array[x])
        
def insertion_sort(arr):
    global count1
    
    for i in range(1, len(arr)):
        for j in range(i, 0, -1):
            count1+=1
            if (arr[j]<arr[j-1]):
                temp=arr[j-1]
                arr[j-1]=arr[j]
                arr[j]=temp 
            else:
                break;
                
       
    return count1

def mergesort(arr):
    
    global count1
    global s
    #print("Value of s",s)
    if (len(arr)<=s):
        return insertion_sort(arr)
        
    
    
    if (len(arr)>1):
        mid=len(arr)//2
        l=arr[:mid]
        r=arr[mid:]
        mergesort(l)
        mergesort(r)
        
        i=j=k=0
        #i is left counter
        #r is right coumter
        #k is arr count 
        while(i<len(l) and j<len(r)): 
            
            count1+=1
            if l[i]<=r[j]: #l is smaller than to right
                arr[k]=l[i]
                i+=1
                
            else:
                arr[k]=r[j]
                j+=1
            k+=1
  
            
        while i<len(l): 
           
            #r no m)re, so l is left and alr sorted
            #all is bigger than the biggest in the arr
            arr[k]=l[i]
            i+=1
            k+=1
            #count+=1
        while j<len(r):
           
            #same for this 
            arr[k]=r[j]
            j+=1
            k+=1
            #count+=1
        
    return count1

        
    
    
        
    
```

```python
#s is the thereshold
merge=[]
insert=[]
size=[]
test=51
step=5
s=0
for i in range(0, test,step): 
    size.append(i)
for i in range(len(size)):
    print(size[i])

for i in range (len(size)):
    random_no=randno(size[i], largest)
    print("size",size[i])
    
    arr1=random_no.copy()
    arr2=random_no.copy()

    count2=0
    count1=0
    count2=insertion_sort(arr1)
    insert.append(count2)
    print("insert",count2)

    count1=0
    count1=mergesort(arr2)
    print("merge",count1)
    merge.append(count1)

intercept_size = None  # Initialize to None, indicating no intercept found yet

# Iterate through the comparison counts and corresponding array sizes
for i in range(len(size)):
    if insert[i] > merge[i]:
        intercept_size = size[i]  # Found the intercept, store the size
        break  # Exit the loop once the intercept is found

if intercept_size is not None:
    print(f"The intercept occurs at array size: {intercept_size}")
else:
    print("No intercept found in the specified range.")

# Create a plot
plt.plot(size, insert, label='Insertion Sort', marker='o')
plt.plot(size, merge, label='Merge Sort', marker='s')

# Highlight the intercept point with a red dot
if intercept_size is not None:
    plt.plot(intercept_size, insert[size.index(intercept_size)], 'ro', label='Intercept')

# Label the axes and add a legend
plt.xlabel('Array Size')
plt.ylabel('Number of Comparisons')
plt.legend()

# Set a title
plt.title('Comparison of Insertion Sort and Merge Sort')

# Show the plot
plt.show() 
```
